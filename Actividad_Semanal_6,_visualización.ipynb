<h4>5. Realiza el conteo de las variables categóricas<h4>

```python
df_2 = df.drop(labels=['ID','GENERO','EDUCACION','ESTADO_CIVIL', 'H_SEP','H_AGO','H_JUL','H_JUN','H_MAY','H_ABR','Y'],axis=1)
df_2.head()
```

<h4>6. Las variables categoricas que podemos encontrar dentro del dataset son:
"GENERO", "EDUCACION", "ESTADO CIVIL" y "Y"<h4>

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df_2)

df_scaled = pd.DataFrame(scaled, columns = df_2.columns)
df_scaled.head()
```

<h4>7. Reduce las dimensiones con PCA, si consideras necesario.<h4>

```python
pcs = PCA()
pcs_t = pcs.fit_transform(df_scaled)
pcs_t
```

**1. Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%**

```python
pca_labels = [f'PC{i + 1}' for i in range(len(df_scaled.columns))]
df_pca = pd.DataFrame({
  '% varianza explicada': np.round(pcs.explained_variance_ratio_,4) * 100,
  '% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_) * 100
  })

df_pca.index = pca_labels

df_pca
```

**2. Indica la importancia de las variables en cada componente**

```python
df_Fe = pd.DataFrame(
  pcs.components_.round(4), 
  columns = df_pca.index,
  index = df_scaled.columns) 

df_Fe.iloc[:,:5]
```
