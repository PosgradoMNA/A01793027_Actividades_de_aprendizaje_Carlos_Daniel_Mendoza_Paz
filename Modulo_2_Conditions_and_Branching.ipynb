**For loop**
Se utiliza para ejecutar un statement un cierto número de veces 

```python
squares = ["Red", "Yellow", "Green"]
```

```python
#Para iterar una lsta
for square in squares:
  print(square)
```

```python
#Cambiar todos los valores de la lista a "White"
for i in range(0,3):
  squares[i] = "White"

print(squares)
```

**Función Enumerate**
Nos ayuda a obtener el indice de los datos, así como los datos también

```python

for i,square in enumerate(squares):
  print(i,square)
  
```

**While loop**

Este trabaja basado en una condición. Mientras la condición no se cumpla el ciclo seguirá ejecutandose.

```python
#Llenaremos una lista con un while, solo dejara pasar los valores que sean "Orange" hacia la nueva lista
squares = ['orange', 'orange', 'purple', 'blue', 'orange']
Newsquares = []
i = 0

while(squares[i] == "orange"):
  Newsquares.append(squares[i])
  i=i+1

print(Newsquares)

```

# **Collecting Arguments**


```python
#*args en una función se usa para pasar de forma opcional un numero variable de argumentos posicionales. Esos argumentos son recibidos como una Tupla
# y este es un parametro opcional por ende se puede o no poner en la función. y se indica con el *

def ArtistNames (*names):
  for name in names:
    print(name)

ArtistNames("Michael Jackson", "AC/DC", "Pink Floyd")
```

# **Scope of a variable**
Podemos tener variables locales o globales, aquellas que sean definidas dentro de una función Python las considerará como una variable local. Aquellas otras que esten declaradas fuera de una función serán consideradas variables globales.
De igual forma puedes crear tu una variable global " global animal = "Gato" "

```python
#Podemos tener variables globales o variables locales
def AddDC(x):
  x = x + "DC" #Aquí x es una variable local por ende no entraria en conflicto si una variable global tiene el mismo nombre
  print(x)
  return (x)

x = "AC" #Aquí x es una variable global
z = AddDC(x) 
```

```python
def Thriller():
  Date = 1982 #Aquí date es una variable local, por ende solo existe en el scope de las variables de esta función
  return(Date)

Date = 2017 #Aquí Date es una variable global, por ende no entra en conflicto con la otra (variable loca) que se llaman igual

print(Thriller())
print(Date)
```
